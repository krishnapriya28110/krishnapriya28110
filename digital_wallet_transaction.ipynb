11 calculate the trend in transaction over the past year by month

```python
transaction_date=pd.to_datetime(df['transaction_date'])
print(transaction_date)

one_year_ago=pd.Timestamp.now()-pd.DateOffset(year=1)
df_last_year=df[df['transaction_date']>=one_year_ago]
#extract year and month for grouping
df_last_year['year-month']=df_last_year['transaction_date'].dt.to_period('M')
#count transaction per month
monthly_trend=df_last_year.groupby('year-month').size()
print(monthly_trend)
```

12 display a graph of trend in transaction over the past year by month

```python
one_year_ago=pd.Timestamp.now()-pd.DateOffset(year=1)
df_last_year=df[df['transaction_date']>=one_year_ago]
#extract year and month for grouping
df_last_year['year-month']=df_last_year['transaction_date'].dt.to_period('M')
#count transaction per month
monthly_trend=df_last_year.groupby('year-month').size()
print(monthly_trend)
#plotting the trend
plt.figure(figsize=(10,5))
monthly_trend.plot(marker='o')
plt.title('transaction trend over the past year')
plt.xlabel('month')
plt.ylabel('number of transaction')
plt.grid()
plt.show()
```

The monthly transaction trend analysis allows you to identify seasonal patterns, evaluate month-on-month changes, and pinpoint gaps in transactions. These insights can inform strategic planning, optimize marketing efforts, and enhance inventory management based on customer behavior.

14 display a bar chart related to total transaction by product category

```python
import matplotlib.pyplot as plt

category_total = df.groupby('product_category')['product_amount'].sum().reset_index().head(15)

# Identifying the maximum value
max_value = category_total['product_amount'].max()
max_index = category_total[category_total['product_amount'] == max_value].index[0]

# Creating a bar chart
plt.figure(figsize=(10,6))
colors = ['skyblue'] * len(category_total) 
colors[max_index] = 'orange'  

plt.bar(category_total['product_category'], category_total['product_amount'], color=colors)
plt.title('Total Transactions by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Amount')
plt.xticks(rotation=50)
plt.grid(axis='y')

# Adding a text label for the highest value
plt.text(max_index, max_value + 5, f'{max_value}', ha='center', color='black', fontsize=10)

# Display the plot
plt.tight_layout( ) 
plt.show()
```

The bar chart effectively showcases the total transaction amounts across different product categories, highlighting the category with the highest sales in orange for emphasis. This visualization helps identify key revenue drivers and informs inventory and marketing strategies, allowing businesses to focus on popular categories to maximize sales potential.
